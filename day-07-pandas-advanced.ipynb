```python
import pandas as pd
import numpy as np

#apply() function lets you to apply function on each row and column it works like loop
data = {
    "name": ["Rakhi", "John", "Sara", "Tom"],
    "salary": [45000, 62000, 38000, 71000]
}
df = pd.DataFrame(data)
#print(df)

def calculate_tax(salary):
    return salary*0.10
df["tax"] = df["salary"].apply(calculate_tax)
#                ↑
# Takes each value from HERE one by one
# and passes it to calculate_tax function
df["tax2"] = df["salary"].apply(lambda x: x * 0.10)
#     ↑               ↑                        ↑
#New column       take each salary        multiply by 0.10
#called tax2      value one by one        return result
#lambda salary:salary*.10

# 1. Create "salary_in_thousands" column
df["salary_in_thousands"] = df["salary"].apply(lambda x: x/1000)
#    hint: divide salary by 1000

# 2. Create "slab" column
#    if salary > 50000 → "High"
#    if salary <= 50000 → "Low"
#    hint: lambda x: "High" if x > 50000 else "Low"
df["slab"] = df["salary"].apply(lambda x : "High" if x>50000 else "Low")

#print(df)

#Part 2 — String operations on DataFrame
data = {
    "name": ["  rakhi ", "JOHN", "sara", "  TOM  "],
    "city": ["mumbai", "DELHI", "Bangalore", "chennai"]
}
df2 = pd.DataFrame(data)
# .str. lets you apply string functions to entire column
# Remove extra spaces
df2["name"] = df2["name"].str.strip()   # .str. tells Pandas
# Convert to proper case (First letter capital)
df2["name"] = df2["name"].str.title()   # "apply this to each
# Convert city to uppercase
df2["city"] = df2["city"].str.upper()   # string in the column"
# other functions
df2["name"].str.upper()      # RAKHI, JOHN
df2["name"].str.lower()      # rakhi, john
df2["name"].str.title()      # Rakhi, John
df2["name"].str.strip()      # remove spaces
df2["name"].str.len()        # length of each name
df2["name"].str.contains("a") # True/False if contains "a"
df2["name"].str.replace("a", "x") # replace characters

# 1. Print length of each name
#print(df2["name"].str.len())

# 2. Find names containing letter "a"
print(df2[df2["name"].str.contains("a")][["name"]]) #specific column else all column will come 
#df[condition] returns ALL columns, df[condition]["name"]   # returns ONLY name column
#print(df2)

# Table 1 - Employee info
employees = pd.DataFrame({
    "emp_id": [1, 2, 3, 4],
    "name": ["Rakhi", "John", "Sara", "Tom"],
    "dept_id": [101, 102, 101, 103]
})

# Table 2 - Department info
departments = pd.DataFrame({
    "dept_id": [101, 102, 103],
    "dept_name": ["IT", "HR", "Finance"]
})

#print(employees)
#print(departments)
# Merge like SQL JOIN SELECT * FROM employees JOIN departments ON employees.dept_id = departments.dept_id;
merged = pd.merge(employees, departments, on="dept_id")
#print(merged)

df3 = pd.read_csv("https://raw.githubusercontent.com/MainakRepositor/Datasets/master/supermarket_sales.csv")
print(df3.columns.tolist())

# 1. Create new column "Total_after_tax" = Total + Tax 5%
#    hint: use lambda
df3["Total_after_tax"] = df3["Total"].apply(lambda x: x+.05)

# 2. Convert "City" column to uppercase
#    hint: use .str.upper()
df3["City"] = df3["City"].str.upper()

# 3. Find all rows where City contains "an"
#    hint: use .str.contains()
df3[df3["City"].str.contains("an")]
print(df3.head(2))
```
