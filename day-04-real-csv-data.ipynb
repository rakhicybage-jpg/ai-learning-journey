```python
import pandas as pd

# Load real dataset directly from internet!
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
print(df.shape)
#see first 5 rows
print(df.head())
print(df.columns.tolist())
#check missing values
print(df.isnull().sum())
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df = df.drop("Cabin", axis=1)
print(df.isnull().sum())
print(df["Survived"].value_counts())
print(df.groupby("Sex")["Survived"].mean())
print(df.groupby("Pclass")["Survived"].mean())
print(df.groupby("Survived")["Age"].mean())
print("Most expensive fare:", df["Fare"].max())
print("Cheapest fare:", df["Fare"].min())
print("Average fare:", round(df["Fare"].mean(), 2))
# Who paid the highest fare and did they survive?
print(df[df["Fare"] == df["Fare"].max()])
#min fare
print(df[df["Fare"] == 0] [["Name", "Age", "Sex", "Pclass", "Survived"]])
```
