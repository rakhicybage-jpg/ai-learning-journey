```python
import numpy as np
import pandas as pd

data = {
    "title":  ["PHP Tutorial", "AI Basics", "WP Guide", "Python Tips", "ML Intro"],
    "author": ["Rakhi", "John", "Rakhi", "Sara", "John"],
    "views":  [1500, 3200, 870, 450, 2100],
    "likes":  [230, 400, 110, 55, 320]
}

df = pd.DataFrame(data)
print(df)
```

```python
#info of data
print(df.info())
```

```python
#Picking single column
print(df['views'])

#picking multiple column
print(df[['views','likes']])

#Single bracket [] → just the data, like a simple list
#Double bracket [[]] → full table format, like MySQL result
```

```python
#filtering like we do in sql
print(df[df['views']>1000])

# Show only Rakhi's posts
print(df[df["author"] == "Rakhi"])

# Rakhi's posts AND views above 1000
print(df[(df["author"] == "Rakhi") & (df["views"] > 1000)])
#Exactly like sql query : SELECT * FROM posts WHERE author = 'Rakhi' AND views > 1000;
#likes are less than 200
print(df[df["likes"] < 200])
```

```python
#Sorting like Sql Orderby
print(df.sort_values("views", ascending=False)) #highest to lowest exactly like sql query : SELECT * FROM posts ORDER BY views DESC;
#in pandas there is no descending only ascending following by True/false
```

```python
#Basic Stats exactly like sql SELECT SUM(views), AVG(views), MAX(views), MIN(views) FROM posts;
print("Sum: ", df["views"].sum())
print("Average: ", df["views"].mean())
print("Max: ", df["views"].max())
print("Min: ", df["views"].min())
```

```python
#value_counts() (like SQL GROUP BY) SELECT author, COUNT(*) FROM posts GROUP BY author;
print(df["author"].value_counts()) # sort by highest

# Count how many posts each author has
# but sort by name A-Z sort_index()
#value_counts() → sort by count
#sort_index() → sort by name A-Z
print(df["author"].value_counts().sort_index()) 
```

```python
print(df.describe())  #give all calculation in one shot
#std : standard deviation (first find distance of each value from average value, square each distance value, take mean of those values, than squareroot of that mean value, )
```

```python
#Adding new column
df["engagement"] = df["views"]  + df["likes"]
print(df)
```

```python
# Create a new column "views_in_thousands"
# hint: divide views by 1000
df["views_in_thousands"] = df["views"]/1000
#print(df)

#Deleting a column
# Drop the views_in_thousands column
df = df.drop("views_in_thousands", axis=1)
print(df)
#axis=1 means column — just remember: axis=0 → row,  axis=1 → column
```

```python
#Handling missing values
# Create data with missing values (None = missing)
data2 = {
    "name":   ["Rakhi", "John", None, "Sara"],
    "views":  [1500, None, 870, 450],
    "likes":  [230, 400, None, 55]
}
df2 = pd.DataFrame(data2)
#print(df2)
# How many missing values in each column?
#print(df2.isnull().sum())

# Fill missing number with average
df2["views"] = df2["views"].fillna(df2["views"].mean())
df2["likes"] = df2["likes"].fillna(round(df2["likes"].mean(),2))

# Fill missing name with "Unknown"
df2["name"] = df2["name"].fillna("Unknown")

print(df2)

print(df2.isnull().sum())
```
