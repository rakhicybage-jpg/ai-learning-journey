```python
import pandas as pd

data = {
    "name": ["Yoast SEO", "WooCommerce", "Akismet", "ContactForm7", "Elementor"],
    "active_installs": [5000000, 4000000, None, 1000000, 3000000],
    "rating": [4.5, 4.2, None, 3.8, 4.7],
    "price": ["Free", "Freemium", "Free", None, "Freemium"]
}
df = pd.DataFrame(data)
print(df.shape)
print(df.info())
# Plugins with rating above 4
print(df[df["rating"]>4])
print(df["active_installs"].mean())
#sort by rating highest to lowest
print(df.sort_values("rating", ascending=False))
# Fill missing active_installs with average
df["active_installs"] = df["active_installs"].fillna(df["active_installs"].mean())
print(df)
df["rating"] = df["rating"].fillna(df["rating"].mean())
#print(df)
df["price"] = df["price"].fillna("Free")
#print(df)
#add new column
df["popularity"] = df["active_installs"]+df["rating"]
#print(df)
#drop columns
df = df.drop("popularity", axis=1) #axis=1 (column), axis=0 (row)
#print(df)
#count free vs paid
print(df["price"].value_counts()) #counts how many times each unique value appears in a column, like group by ijn sql and array_count_values in php
#print(df)
print(df.describe()) #includes only number column 
print(df.describe(include="all")) #includes text column also
pd.options.display.float_format = '{:.2f}'.format
print(df.describe())
```
