```python
def calculateBMI(weight, height):
    bmi = weight / (height ** 2)
    return round(bmi, 2)
print(calculateBMI(66, 1.75))
print(calculateBMI(90, 1.80))
```

```python
post = {
    "title": "My First AI Blog",
    "author": "Rakhi",
    "views": 1500,
    "published": True
}

print(post["title"])

post["views"] = 2000
print(post["views"])

for key, value in post.items(): # just same we use $key=>$value in php here .items() is python inbuilt function to get all key value at once
    print(key, "→", value)
```

```python
#trying some python super power techniques
scores = [76,89,34,23,56,67,78,89,90]
unique_score = set(scores) # if want unique scores than use inbuilt set() function just like we use array_unique() in php
passed = [score for score in unique_score if score >=60]
print(passed)
```
