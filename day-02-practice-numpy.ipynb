```python
import numpy as np
salaries = np.array([25000, 42000, 38000, 55000, 29000, 61000, 47000])

# 1. Print all salaries above 40000
salary = salaries[salaries > 40000]
print(salary)
# 2. Print average salary
print(round(np.mean(salaries),2))
# 3. Print highest and lowest salary
print(np.max(salaries))
print(np.min(salaries))
# 4. Give everyone a 15% raise and print new salaries
inc_salary = salaries + (salaries*15/100)
print(inc_salary)
# 5. Print salaries between 30000 and 50000
print(salaries[(salaries > 30000) & (salaries < 50000)])
```

```python
posts = np.array([
    [1500, 230, 45],
    [870,  110, 23],
    [450,   55, 12],
    [2300, 400, 89],
    [120,   10,  3],
    [3100, 500, 120],
    [670,   80, 18],
])

# 1. Shape
print("Shape:", posts.shape)

# 2. Posts with views above 1000
print(posts[posts[:, 0] > 1000])

# 3. Average views
print(round(np.mean(posts[:, 0]), 2))

# 4. Everyone's likes (column 1)
print(posts[:, 1])

# 5. Post with maximum views
idx = np.argmax(posts[:, 0]) #np.argmax find the highest value index
print(posts[idx])
```

```python
sales = np.array([
    [15000, 18000, 21000],
    [9000,  11000, 13000],
    [22000, 25000, 28000],
    [7000,   8000,  9000],
])

# 1. Print shape
print("Shape:", sales.shape)

# 2. Employee 3's all sales (row index 2)
print("Employee 3 sales:", sales[2])

# 3. Everyone's January sales (column 0)
print("January sales:", sales[:, 0])

# 4. Employee 2's February sales (row 1, column 1)
print("Employee 2 February:", sales[1][1])

# 5. Average of all sales
print("Average sales:", round(np.mean(sales), 2))
```
